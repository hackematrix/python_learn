**同时表示多条**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
sin_y=np.sin(x)

#绘制正弦曲线
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(9,8),dpi=80)
plt.plot(x,sin_y,linestyle='-',marker='.')

#绘制余弦曲线
cos_y=np.cos(x)
plt.plot(x,cos_y,linestyle='-',marker='.')

plt.show()
```

**subplot**

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

#设置两个子图
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.show()
```

**使用bar和barh绘制柱状图**

**柱状叠加图**

```python
import numpy as np
import matplotlib.pyplot as plt

countries=['USA','India','China','Russia','Germany']
bronzes=np.array([38,17,26,19,15])
silvers=np.array([37,23,18,18,10])
golds=np.array([46,27,26,19,17])

ind=[x for x,_ in enumerate(countries)]

plt.bar(ind,golds,width=0.5,label='golds',color='gold',bottom=silvers+bronzes)
plt.bar(ind,silvers,width=0.5,label='silvers',color='silver',bottom=bronzes)
plt.bar(ind,bronzes,width=0.5,label='brozes',color='#CD853F')

plt.xticks(ind,countries)
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.legend(loc='upper right')
plt.title('2019 Olympics Top Scores')

plt.show()

```

**3D线图**

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()

ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(20*z)
y=z*np.cos(20*z)

ax.plot3D(x,y,z,'grey')
ax.set_title('3D line plot')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(20*z)
y=z*np.cos(20*z)
c=x+y

ax.scatter3D(x,y,z,c=c)
ax.set_title('3D scatter plot')
plt.show()

```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure()
ax=plt.axes(projection='3d')

x=np.outer(np.linspace(-2,2,30),np.ones(30))
y=x.copy().T
z=np.cos(x**2+y**2)

ax.plot_surface(x,y,z,cmap='viridis',edgecolor='none')
ax.set_title('3D Surface Plot')

plt.show()
```

**综合**

```python
import matplotlib.pyplot as plt

data={'apple':10,'orange':15,'lemon':5,'lime':20}
names=data.keys()
values=data.values()
print(names)
print(values)

fig,axs=plt.subplots(1,3,figsize=(9,3),sharey=True)
axs[0].bar(names,values)
axs[1].scatter(names,values)
axs[2].plot(names,values)
fig.suptitle('catagrical plot')
```
